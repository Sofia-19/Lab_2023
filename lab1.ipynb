# Лабораторна робота 1

Дана послідовність з 0 та 1, яка була отримана зміною кожної цифри початкової послідовності(яка також була з 0 і 1) з ймовірністю 0.5 для кожної цифри, незалежно одна від одної. 

Потрібно знайти оптимальну стратегію для відшукання суми всіх цифр початкової послідовності. 

Штрафною функцією буде різниця між сумами 2 послідовностей в квадраті.

# Хід роботи

```python
import numpy as np
import itertools
```

#### Створимо функцію для генерації всіх можливих бінарних (0,1) послідовностей довжини n:

```python
def get_seq(n):
    k = [0,1]
    q = []
    for x in itertools.product(k, repeat=n):
        q.append(x)
    return q
```

#### Задамо тестові вхідні дані:

```python
new_seq = [0,1,1,1] #якась послідовність, яка була утворена зміною кожної цифри початкової послідовності (Х)
n = len(new_seq) #визначимо довжину цієї послідовності
```

#### Визначимо параметри для формування Баєсiвської задачi:

```python
K = get_seq(n) #Визначимо всі можливі стани 
old_seq = K #Визначимо множину розв'язків D (K=D)
```

```python
old_seq
```

#### Задамо pXK : X × K → R – ймовiрнiсний розподiл на множинi пар (x, k).

```python
p_xk = (1/(2**n))*(1/(2**n))
```

#### Шукаємо оптимальну стратегію, тобто функцію q : X → D, яка мiнiмiзує баєсiвський ризик 

q∗(x) ∈ argmin_d (sum_k(pK|x(k)W(k, d))

d∈D
k∈K

#### Штрафною функцією буде різниця між сумами 2 послідовностей в квадраті: 
    
W(k,d) = (sum(k)-sum(d))**2

```python
sum_se = [] #Для збереження суми штрафної функції при різних к та d
old_seq_sum = [] #Для збереження суми послідовності
for j in range (0, len(old_seq)): #перебираємо всі можливі значення початкової послідовності (D)
    print("\033[1mIf Old seq: ",old_seq[j], "\tOld seq sum:", sum(old_seq[j]),"\033[0m\n")
    se = [] #Для збереження штрафної функції при різних к та фіксованого d
    for i in range(0,len(K)): #перебираємо всі можливі к
        print("k seq: ",K[i], "\t\t k seq sum:", sum(K[i]))
        se.append(p_xk*(sum(old_seq[j])-sum(K[i]))**2)# рахуємо штрафну функцію при певному k   
        print(p_xk,"*(",sum(old_seq[j]),"-",sum(K[i]),")**2 = ", se[-1],"\n")
    old_seq_sum.append(sum(old_seq[j])) 
    sum_se.append(sum(se))#Рахуємо sum_k(pK|x(k)W(k, d) і поміщаємо в список
    print("\033[1mSquared error: ", sum_se[-1],'\033[0m')
    print("\n")
```

#### Тобто обтимальною буде стратегія, за якою в якості початкової послідовності треба обирати таку послідовність, сума цифр якої дає найменше значення штрафної функції  

argmin_d(sum_k(pK|x(k)W(k, d))

#### Тоді оптимальна стратегія для наших тестових даних:

```python
print("Optimal q: ", sum(old_seq[np.argmin(sum_se)]))    

print(f"Optimal strategy: Choose sequense {0,1}**{n} with sum {sum(old_seq[np.argmin(sum_se)])}")
```
